```python
# Load by directory (preferred)
result_base = os.path.expandvars('$SCRATCH/heptrkx/results')
```

```python
result_names = ['resmpnn_noW3', 'resmpnn_noW4', 'resmpnn_noW5', 'resmpnn_noW6']
```

```python
configs = []
summaries = []

for i, result_name in enumerate(result_names):
    result_dir = os.path.join(result_base, result_name)

    configs.append(load_config_dir(result_dir))
    summaries.append(load_summaries(configs[i]))
```

```python
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(ncols=2, nrows=2, figsize=(16, 12))

# Color cycle
prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']

# Plot the results
for name, summary, color in zip(result_names, summaries, colors):
    ax0.plot(summary.epoch, summary.train_loss, c=color, linestyle='dashed', label=name+' train')
    ax0.plot(summary.epoch, summary.valid_loss, c=color, label=name+' validation')
    ax1.plot(summary.epoch, summary.valid_acc, c=color, label=name+' validation')
    
    if 'l1' in summary:
        ax2.plot(summary.epoch, summary.l1, c=color, label=name)
        ax3.plot(summary.epoch, summary.l2, c=color, label=name)

ax0.set_xlabel('Epoch')
ax0.set_ylabel('Loss')
ax0.legend(loc=0)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
#ax1.set_ylim(bottom=0, top=1)
ax1.legend(loc=0)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('L1 norm')
ax2.legend(loc=0)
ax3.set_xlabel('Epoch')
ax3.set_ylabel('L2 norm')
ax3.legend(loc=0)

plt.tight_layout()
```
